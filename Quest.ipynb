```python
def convert(value):
    if value == 'TELD':
        return 1
    elif value == 'IELD':
        return 0
```

```python
df_ROS_data['수술기법'] = df_ROS_data['수술기법'].apply(convert)
```

```python
target = df_ROS_data['수술시간']
features = df_ROS_data.drop(columns=['수술시간'])
target.shape, features.shape
```

```python
from sklearn.model_selection import train_test_split
features_train, features_test, target_train, target_test = train_test_split(features, target, random_state=111)
features_train.shape,features_test.shape, target_train.shape, target_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
```

```python
model = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
hyper_parmas = {'min_samples_leaf' : [2,5],'max_depth' : [2,5],'min_samples_split' : [2,5]}
```

```python
from sklearn.metrics import f1_score, make_scorer
```

```python
scoring = make_scorer(f1_score)
```

```python
grid_search = GridSearchCV(model, param_grid=hyper_parmas, cv=3, verbose=1, scoring=f1_score)
```

```python
grid_search.fit(features, target)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_score_, grid_search.best_params_
```

```python
best_model = grid_search.best_estimator_
best_model
```

```python
df_ROS_Redata = df_ROS[['수술시간', '성별','고혈압여부','수술기법', '전방디스크높이(mm)','후방디스크높이(mm)','Modic change','디스크단면적']]
```

```python
df_ROS_Redata['수술기법'] = df_ROS_Redata['수술기법'].apply(convert)
```

```python
df_ROS_Redata.isnull().sum()
```

```python
# 모델을 사용하여 결측치를 채우는 함수 정의
def fill_missing_values(row):
    # 결측치를 채우고자 하는 열의 값을 가져옴
    if pd.isnull(row['수술시간']):
        # 결측치인 경우, 모델을 사용하여 예측
        features = row[['성별','고혈압여부','수술기법', '전방디스크높이(mm)','후방디스크높이(mm)','Modic change','디스크단면적']].values.reshape(1, -1)
        predicted_value = best_model.predict(features)
        return predicted_value[0]
    else:
        # 결측치가 아닌 경우, 그대로 반환
        return row['수술시간']
```

```python
# 결측치를 채우고자 하는 열에 함수 적용
df_ROS_Redata['수술시간'] = df_ROS_Redata.apply(fill_missing_values, axis=1)
df_ROS_Redata['수술기법'] = df_ROS_Redata.apply(fill_missing_values, axis=1)
df_ROS_Redata['디스크단면적'] = df_ROS_Redata.apply(fill_missing_values, axis=1)
```

```python
# 결측치가 채워진 데이터프레임 확인
print(df_ROS_Redata.isnull().sum())
```
